# Playing with sigmoid function

An important element of the neural network is the activation function. The weight and bias of a neuron shape the output of the activation to produce the desired output.

```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

The equation of the function looks like

![](https://latex.codecogs.com/svg.image?\frac{1}{1&plus;e^{-(wx&plus;b)}})

```python
def sigmoid(x,weight,bias):
    y = 1 / (1 + math.exp(-(x*weight + bias)))
    return y
```

```python
def buildOneCurve(Xs,w,bias):
    onecurve=[]
    for x in Xs:
        y = sigmoid(x,w,bias)
        onecurve.append(y)
    return onecurve
```

```python
results=[]
Xs=range(-5,5)
results.append(buildOneCurve(Xs,0.5,0))
results.append(buildOneCurve(Xs,1,0))
results.append(buildOneCurve(Xs,1.5,0))
results.append(buildOneCurve(Xs,2,0))
```

```python
plt.plot(Xs,results[0],Xs,results[1],Xs,results[2],Xs,results[3])
plt.axis((-5,5,0,1))
plt.show()
```

Bias does shift the sigmoid curve, which allows values other than 0.5 when x is near 0

```python
results=[]
results.append(buildOneCurve(Xs,1,1))
results.append(buildOneCurve(Xs,1,0.5))
results.append(buildOneCurve(Xs,1,1.5))
results.append(buildOneCurve(Xs,1,2))
plt.plot(Xs,results[0],Xs,results[1],Xs,results[2],Xs,results[3])
plt.axis((-5,5,0,1))
plt.show()
```
